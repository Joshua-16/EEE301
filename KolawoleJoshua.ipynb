```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
V =  np.arange(158)
f = np.sqrt(V/(4* (np.pi**2) * (5 * 10**-9) * (1.6 * 10** -19)))
```

```python
plt.xlabel('Voltage')
plt.ylabel('Frequency')
plt.title('Plot of Frequency against Voltage')
plt.plot(V, f)
plt.savefig('plot.png')
```
